```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
senior_fentanyl_allyears = pd.read_csv("AllYearSeniorFentanyl.csv")
```

```python
senior_fentanyl_allyears
```

```python
plt.rcParams["figure.figsize"] = (20,10)
```

```python
plt.plot(senior_fentanyl_allyears.year, senior_fentanyl_allyears.rate)
plt.title("Crude Rate Per 100K of Fentanyl Overdose Date Rates for Seniors vs Year")
plt.xlabel("year")
plt.ylabel("Crude Rate Per 100K")
```

```python
teenager_fentanyl_allyears = pd.read_csv("AllTeenagersFentanylOverdoseRate.csv")
```

```python
teenager_fentanyl_allyears
```

```python
plt.plot(teenager_fentanyl_allyears.year, teenager_fentanyl_allyears.rate)
plt.title("Crude Rate Per 100K of Fentanyl Overdose Date Rates for Teenagers vs Year")
plt.xlabel("year")
plt.ylabel("Crude Rate Per 100K")
```

```python
plt.plot(teenager_fentanyl_allyears.year, teenager_fentanyl_allyears.rate)
plt.title("Crude Rate Per 100K of Fentanyl Overdose Date Rates vs Year")
plt.xlabel("year")
plt.ylabel("Crude Rate Per 100K")
plt.plot(senior_fentanyl_allyears.year,senior_fentanyl_allyears.rate)
plt.legend(["Teenagers", "Seniors"])
```
