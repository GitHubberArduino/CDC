```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from sympy import *
import lmfit
from lmfit import Parameters, minimize
from lmfit.printfuncs import report_fit

```

```python
x = [2016, 2017, 2018, 2019, 2020, 2021]
fentanylall = [0.31, 0.228, 0.254 ,0.232, 0.232, 0.242 ]
opioidall = [0.278,0.23, 0.208, 0.192, 0.2, 0.204]
```

```python
plt.rcParams["figure.figsize"] = (30,20)
plt.rcParams.update({'font.size': 30})
```

```python
plt.plot(x, fentanylall)
plt.plot(x, opioidall)
plt.title("Gini Index for Lorenz Curves of Crude Deaths for All Ages")
plt.xlabel("Year")
plt.ylabel("Gini Index")
plt.legend(["Fentanyl", "Opioid"], loc='center right', title='Region')
```
