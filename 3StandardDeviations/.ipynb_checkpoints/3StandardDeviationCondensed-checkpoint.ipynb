```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from sympy import *
import lmfit
from lmfit import Parameters, minimize
from lmfit.printfuncs import report_fit

    
```

```python
def fixdf(df):
    df.columns = df.columns.str.strip()
    df = df.drop("Gender", axis = 1)
    df.rename(columns = {"Gender Code": "Gender", "Year Code": "Gender Code"}, inplace = True)
    return df
```

```python
def threestandarddev(df, color):
    plt.rcParams["figure.figsize"] = (30,30)
    gdf = pd.DataFrame({"value": [2013, 2013.5, 2014, 2014.5,2015, 2015.5, 2016, 2016.5, 2017, 
                              2017.5, 2018, 2018.5, 2019, 2019.5, 2020, 2020.5, 2021]})
    df.rename(columns = {"year": "Year", "rate": "rate"}, inplace = True)
    df['Year'] = df['Year'].astype(float)
    dfexceptlasttwoyears = df[(df["Year"]!= 2021.0) & 
                (df["Year"] != 2020.0)]
    dflasttwo = df[(df["Year"]!= 2019.0) & 
                (df["Year"] != 2018.0) & (df["Year"] != 2017.0)&(df["Year"] != 2016.0)&(df["Year"] != 2014.0)&(df["Year"] != 2015.0)&(df["Year"] != 2013.0)]
    x1 = dfexceptlasttwoyears['Year']
    y1 = dfexceptlasttwoyears['rate']
    x2 = dflasttwo['Year']
    y2 = dflasttwo['rate']
    coeff = np.polyfit(x1, y1, 2)#findcoefficients
    yfit = np.polyval(coeff,x1)#solveforybasedonx
    residual = np.sum((y1  - yfit)**2)#determinedistancebetweenpointandfit
    std = np.sqrt(residual/5)
    plt.plot(gdf, np.polyval(coeff,gdf),color= color)
    plt.scatter(x1,y1, color= color, s=500 )
    plt.scatter(x2, y2, color= color, s=500) 
    plt.scatter(x2, np.polyval(coeff,x2), s=500, edgecolors=color, facecolors='none')
    plt.plot(gdf, np.polyval(coeff,gdf)+3*std, color=color, linestyle='dashed')
    plt.plot(gdf, np.polyval(coeff,gdf)-3*std, color=color, linestyle='dashed')
```

```python
def plot(df, color, drugtype):
    if drugtype == 'fentanyl':
        year = [2015,2016,2017,2018,2019,2020,2021]
    else:
        year = [2013,2014,2015,2016,2017,2018,2019,2020,2021]
    x = year
    y = df['rate']
    plt.plot(x,y, color = color)
```

```python
dfexhaustive = pd.read_csv("CollectiveDataFrameOpioidsandMeth.csv")
```

```python
def groupingdataframe(df):
    grouped_df = df.groupby('age')

    df1524 = pd.DataFrame()  # Empty DataFrame for group '15-24'
    df2534 = pd.DataFrame()  # Empty DataFrame for group '25-34'
    df3544 = pd.DataFrame()  # Empty DataFrame for group '35-44'
    df4554 = pd.DataFrame()  # Empty DataFrame for group '45-54'
    df5564 = pd.DataFrame()  # Empty DataFrame for group '55-64'
    df6574 = pd.DataFrame()  # Empty DataFrame for group '65-74'
    for group_name, group_df in grouped_df:
        if group_name == "15-24 years":
            df1524 = df1524.append(group_df)
            df1524 = df1524.sort_values(by='year', ascending=True)

        elif group_name == "25-34 years":
            df2534 = df2534.append(group_df)
            df2534 = df2534.sort_values(by='year', ascending=True)
        elif group_name == "35-44 years":
            df3544 = df3544.append(group_df)
            df3544 = df3544.sort_values(by='year', ascending=True)
        elif group_name == "45-54 years":
            df4554 = df4554.append(group_df)
            df4554 = df4554.sort_values(by='year', ascending=True)
        elif group_name == "55-64 years":
            df5564 = df5564.append(group_df)
            df5564 = df5564.sort_values(by='year', ascending=True)
        elif group_name == "65-74 years":
            df6574 = df6574.append(group_df)
            df6574 = df6574.sort_values(by='year', ascending=True)
    dataframes = [df1524, df2534,df3544,df4554,df5564,df6574]
    return dataframes

```

```python
all1524 = groupingdataframe(dfexhaustive)[0]
all2534 = groupingdataframe(dfexhaustive)[1]
all3544 = groupingdataframe(dfexhaustive)[2]
all4554 = groupingdataframe(dfexhaustive)[3]
all5564 = groupingdataframe(dfexhaustive)[4]
all6574 = groupingdataframe(dfexhaustive)[5]
```
