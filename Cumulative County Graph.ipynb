```python
lower_limit = 0
upper_limit = 80
num_pts = 5000
x = np.linspace(lower_limit, upper_limit, num_pts)
```

```python
def asdf(row):
    row['checky'] = 2.60827798 - 1.89861475*row['rate'] + 0.42611547*(row['rate']**2) - 0.01305627*(row['rate']**3) - 6.7668e-5*(row['rate']**4) + 8.9682e-6*(row['rate']**5) - 1.4051e-7*(row['rate']**6) + 6.8144e-10*(row['rate']**7)
    return row['checky']
```

```python
originalCoefs = [params3["c0"].value,params3["c1"].value,params3["c2"].value,params3["c3"].value,params3["c4"].value,params3["c5"].value,params3["c6"].value,params3["c7"].value]
```

```python
originalCoefs
```

```python
diffCoefs = [originalCoefs[1],2*originalCoefs[2],3*originalCoefs[3],4*originalCoefs[4],5*originalCoefs[5],6*originalCoefs[6],7*originalCoefs[7]]
```

```python
def prime(row):
    row['rate'] = diffCoefs[0] + diffCoefs[1] * row['rate']  + diffCoefs[2] * (row['rate']**2) + diffCoefs[3] * (row['rate']**3) + diffCoefs[4] * (row['rate']**4) + diffCoefs[5] * (row['rate']**5) + diffCoefs[6] * (row['rate']**6)
    return row['rate']
```

```python
df['yprime'] = df.apply(prime,axis=1)
```
