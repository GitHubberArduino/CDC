```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
countyareas = pd.read_csv("areaofcaliforniacounties.txt")
countypopulation2016 = pd.read_csv("CaPopulations2016.csv")
countypopulation2021 = pd.read_csv("PopulationofCA2021.csv")
fentanyl2016 = pd.read_csv("FentanylAll2016.csv")
fentanyl2021 = pd.read_csv("NewFentanyl2021AllAges.csv")
plt.rcParams["figure.figsize"] = (30,20)
plt.rcParams.update({'font.size': 30})
```

```python
def delta(areas,pop1,pop2,rate1,rate2):
    areas["county"] = areas["county"] + " County"
    pop1 = pop1.rename(columns={'population': 'pop1'})
    pop2 = pop2.rename(columns={'population': 'pop2'})
    populations1 = pd.merge(areas,pop1)
    populations2 = pd.merge(areas, pop2)
    combinedpop = pd.merge(populations1,populations2)
    combinedpop['density1'] = combinedpop['pop1']/combinedpop['area']
    combinedpop['density2'] = combinedpop['pop2']/combinedpop['area']
    combinedpop['subdensity'] = combinedpop['density2'] - combinedpop['density1']
    rate1['county'] = rate1['county'] + " County"
    rate1 = rate1.rename(columns={'rate': 'rate1'})
    rate2 = rate2.rename(columns={'rate': 'rate2'})
    rate2['county'] = rate2['county'] + " County"
    rate2 = rate2.drop("num1", axis = 1)
    rate2 = rate2.drop("num2", axis = 1)
    rate2 = rate2.drop("num3", axis = 1)
    rate2 = rate2.drop("deaths", axis = 1)
    combineddf = pd.merge(combinedpop, rate1)
    combineddf2 = pd.merge(combineddf, rate2)
    combineddf2['subrate'] = combineddf2['rate2'] - combineddf2['rate1']
    combineddf2['delta'] = combineddf2['subrate'] / combineddf2['subdensity']
    x = combineddf2['density2']
    y = combineddf2['delta']
    plt.xlabel("Density of County")
    plt.ylabel("Delta")
    plot = plt.scatter(x,y)
```

```python
delta(countyareas,countypopulation2016,countypopulation2021,fentanyl2016,fentanyl2021)
plt.title("Density of County vs Delta, Fentanyl")
```
