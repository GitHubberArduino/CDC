```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
CAquarter = pd.read_csv("CAQuarterlyanyopioid.csv")
```

```python
CAquarter.sort_values("year", axis=0, ascending=True,inplace=True, na_position='first')
```

```python
CAquarter
```

```python
plt.rcParams["figure.figsize"] = (20,10)
```

```python
plt.plot(CAquarter.year, CAquarter.rate)
plt.title("Crude Rate Per 100K of Any Opioid Overdose Date Rates for All Ages by Quarter Year")
plt.xlabel("year")
plt.ylabel("Crude Rate Per 100K")
```
